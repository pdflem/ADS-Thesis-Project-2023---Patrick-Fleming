# Adding studies of UU students

```python
# Data with student studies
new_studies_file = 'sponsor.xlsx'
xl = pd.ExcelFile(new_studies_file)
studies = xl.parse("Werkblad 1", header=None, names=['username', 'dossier_id', 'sponsor'])

# Drop first row
studies.drop(index=studies.index[[0,1]], axis=0, inplace=True)
#studies.reset_index()

# remove column 'username' as Sollicity have verified that it is for admin purposes only and poses no useful information
studies.drop(studies.columns[[0]], axis=1, inplace=True)
studies.rename(columns={"sponsor": "faculty"}, inplace=True)
studies = studies.reset_index(drop=True)
studies
```

```python
# new data with groups

new_groups_file = 'UU Careercheck 20230504.xlsx'

xls = pd.ExcelFile(new_groups_file)
new_df1 = pd.read_excel(xls, 'remarks')
new_df2 = pd.read_excel(xls, 'questionnaires in system')
new_df3 = pd.read_excel(xls, 'questions in system')
new_df4 = pd.read_excel(xls, 'groups in questionnaires in sys')
```

```python
# questioninlevel1subgroup contains the question_ids so renaming
new_df4.rename(columns={"questioninlevel1subgroup": "question_id"}, inplace=True)

# replace question.id_ part of values so that we only have the ID values for ease of merging
new_df4['question_id'] = new_df4['question_id'].str.replace('question.id_', '')

Filt_df4 = new_df4.drop(['level2subgroup', 'level2subgroup_id', 'questioninlevel2subgroup'], axis=1)
Filt_df4
```

```python
# Replace null value with NAN
Filt_df4['question_id'] = Filt_df4['question_id'].replace(r'\N', np.nan)
Filt_df4.question_id.fillna(Filt_df4.questioninmaingroup_id, inplace=True)

# convert question_id float values to int and then to string
Filt_df4['question_id'] = Filt_df4['question_id'].astype(int)
Filt_df4['question_id'] = Filt_df4['question_id'].astype(str)

# droping questioninmaingroup_id and questioninmaingroup
Filt_df4 = Filt_df4.drop(['questioninmaingroup_id', 'questioninmaingroup'], axis=1)

# replace question.id_ part of values so that we only have the ID values for ease of merging
#Filt_df4['question_id'].str.replace('question.id_', '')
Filt_df4
```

Every level1subgroup row appears to contain more than one row per question but the it still links to the same type of result so removing these extra group values e.g. ik houd van (e.g. Social)

```python
# Search for specific words in level1subgroup column and replace with NaN
Filt_df4.loc[Filt_df4['level1subgroup'].str.contains('ik houd van', case=False, na=False), 'level1subgroup'] = np.nan
Filt_df4.loc[Filt_df4['level1subgroup'].str.contains('ik kan', case=False, na=False), 'level1subgroup'] = np.nan
Filt_df4.loc[Filt_df4['level1subgroup'].str.contains('ik ben', case=False, na=False), 'level1subgroup'] = np.nan
Filt_df4.loc[Filt_df4['level1subgroup'].str.contains('taxonomy.onet.cmr.select_highpoint', case=False, na=False), 'level1subgroup'] = np.nan

# Print the updated DataFrame
Filt_df4
```

```python
# Remove rows with NaN values in level1subgroup column
Filt_df4 = Filt_df4.dropna(subset=['level1subgroup'])

#group by 'question.id' and aggregate unique values in subgroup column
grouped = Filt_df4.groupby('question_id')['level1subgroup'].agg(lambda x: tuple(set(x))).reset_index()

# Find the maximum number of unique subgroup values
max_subgroup_count = grouped['level1subgroup'].apply(lambda x: len(x)).max()

# create new columns for each unique subgroup value
new_columns = [f'level1subgroup{i+1}' for i in range(max_subgroup_count)]
grouped[new_columns] = pd.DataFrame(grouped['level1subgroup'].tolist(), index=grouped.index)

# Drop the original subgroup
grouped = grouped.drop('level1subgroup', axis=1)
grouped
```

one question was found to correspond to two separate results when no other questions had this tendency. The Question was to do with friendlyness and hence more likely to be concerned with 'Social' than 'Artistic'

```python
# Setting question 2232 to only social instead of social and artistic

#condition = grouped['level1subgroup2'] == 'Artistiek'  # Define the condition
#grouped.loc[condition, 'level1subgroup1'] = grouped.loc[condition, 'level1subgroup2']
df_groups = grouped.drop('level1subgroup2', axis=1)
df_groups[df_groups['question_id'] == '2232']
```

```python
groups = df_groups # Shorter variable name for simplicity
groups.rename(columns={"level1subgroup1": "level1subgroup"}, inplace=True)
groups
```

Preprocessing skills (Competencies) to be added to groups data too as it is missing level1subgroup data which is needed for outcome of questions

```python
# Import data with skills
skills_file = 'Competencies.xlsx'
skills = pd.read_excel(skills_file)
#skills

# Fix variable names and subset data to include only necessary variables
skills.rename(columns={"id": "question_id"}, inplace=True)
skills.rename(columns={"maincompetences": "level1subgroup"}, inplace=True)
skills_drop = skills.drop(['description', 'question', 'createdOn', 'lastModified', 'type', 'knowledgeBaseKey', 'subcompetences',  'maincompetencesII'], axis=1)
skills_drop.head(6)

# add skills to groups data
groups['level1subgroup'] = groups['level1subgroup'].replace(r'\N', np.nan)
groups = groups.set_index('question_id')
groups['level1subgroup'] = groups['level1subgroup'].combine_first(skills_drop.set_index('question_id')['level1subgroup'])
groups = groups.reset_index()

groups.tail(37)
```

```python
# Loading modified dataset produced in Samuel's research
new_mod_file = 'uu_norm.xlsx'

xls_mod = pd.ExcelFile(new_mod_file)
df_mod = pd.read_excel(xls_mod, 'Sheet 1')
```

```python
# (Ensuring is in) converting to string data type
df_mod['question_id'] = df_mod['question_id'].astype(str)
groups['question_id'] = groups['question_id'].astype(str)

# Merge groups to main dataset
df_mod_groups = pd.merge(df_mod, groups, on='question_id', how='left')
df_mod_groups
```

```python
# Merging students studies to orginal data for Naomi's analysis

# merging (left join) studies to orginal dataframe
df_mod_groups_study = pd.merge(df_mod_groups, studies, on='dossier_id', how='left')
df_mod_groups_study
```

```python
# Removing demonstration dossier_id (accounts) values from orginal dataframe

# data with student demos
dossier_exclude_file = 'uu uva exclude.xlsx'

xl1 = pd.ExcelFile(dossier_exclude_file)
exclude = xl1.parse("Werkblad 1", header=None, names=['dossier_id'])

# Drop first row using drop()
exclude.drop(index=studies.index[[0,1]], axis=0, inplace=True)

# Remove special characters from the dossier_id column
exclude['dossier_id'] = exclude['dossier_id'].str.replace('"', '')

exclude['dossier_id'] = exclude['dossier_id'].astype(int)
df_mod_groups_study['dossier_id'] = df_mod_groups_study['dossier_id'].astype(int)

# Filter and remove dossier_id values from df1 using df2
df_mod_groups_study = df_mod_groups_study[~df_mod_groups_study['dossier_id'].isin(exclude['dossier_id'])]

# Removing unnecessary Variables
#df_mod_groups_study = df_mod_groups_study.drop(['createdOn...5', 'lastModified...6', 'id...9', 'createdOn...13',  'dossierTest_id', 'lastModified...14', 'description', 'knowledgeBaseKey', 'systemName', 'allowEmptyAnswer'], axis=1)
df_mod_groups_study
```
